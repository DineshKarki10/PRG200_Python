# load the required libraries 

```python
import pandas as pd # used for data manipulation and analysis
```

# load the dataset (csv), and checking the dataframe 

```python
df = pd.read_csv('bhatbhateni_sales.csv') # read the CSV file into a DataFrame
print(df.head()) # print the first 5 rows of the DataFrame
print(df.shape) # print the shape of the DataFrame
print(df.columns) # print the column names of the DataFrame
print("Rows:", df.shape[0]) # Number of rows
print("Columns:", df.shape[1]) # Number of columns 
```

List the columns 

```python
list(df.columns) # list the column names of the DataFrame
```

# Data Types, and Structure


```python
print(df.info()) # print the summary information of the DataFrame
print(df.describe(include='all')) # print the statistical summary of the DataFrame
print(df.dtypes) # print the data types of each column in the DataFrame
```

# Detech Data Quality Issues

```python
missing_count = df.isnull().sum() # count the number of missing values in each column
print(missing_count) # print the count of missing values in each column
missing_percent = (missing_count / len(df)) * 100 # calculate the percentage of missing values in each column
print(missing_percent) # print the percentage of missing values in each column  
# Combine both into one DataFrame
missing_summary = pd.DataFrame({
    "Missing Values": missing_count,
    "Percentage (%)": missing_percent.round(2)
})
print(missing_summary)
```

# Duplicated 

```python
duplicate_rows = df.duplicated().sum() # count the number of duplicate rows in the DataFrame

print(f"Number of duplicate rows: {duplicate_rows}") # print the number of duplicate rows in the DataFrame  
```

Check Duplicate Transaction IDs


```python
transaction_duplicates = df["TransactionID"].duplicated().sum() # count the number of duplicate Transaction IDs in the DataFrame

print(f"Repeated Transaction IDs: {transaction_duplicates}")

# View all records with repeated Transaction IDs
df[df["TransactionID"].duplicated(keep=False)].sort_values("TransactionID")
```

Validate Total Amount

```python
expected_total = df["Quantity"] * df["UnitPrice"] # Calculate expected total

incorrect_total = df[df["TotalAmount"] != expected_total] # Find mismatched rows

print("Incorrect TotalAmount rows:", len(incorrect_total))

incorrect_total.head()
```

# Remove Duplicate Rows

```python
df = df.drop_duplicates() # Remove duplicate rows from the DataFrame

print("Duplicates removed successfully.") # print a success message
```

# Verify Duplicate Removal

```python
print("Current Shape:", df.shape) # print the current shape of the DataFrame after removing duplicates  

print("Remaining Duplicates:", df.duplicated().sum()) # print the number of remaining duplicate rows in the DataFrame after removing duplicates
```

# Fill Missing Customer Names

```python
df["CustomerName"] = df["CustomerName"].fillna("Unknown") # fill missing values in the "CustomerName" column with "Unknown"
```

# Fill Missing Product Categories

```python
# Fill missing categories using ProductName
df["ProductCategory"] = (
    df.groupby("ProductName")["ProductCategory"]
      .transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else "Unknown"))
)
```

# Fill Missing Unit Price

```python
# Fill missing UnitPrice using TotalAmount and Quantity
df["UnitPrice"] = df["UnitPrice"].fillna(
    df["TotalAmount"] / df["Quantity"]
)
```

# Fill Missing Payment Method

```python
df["PaymentMethod"] = df["PaymentMethod"].fillna("Unknown") # fill missing values in the "PaymentMethod" column with "Unknown"  
```

# Verify Missing Values

```python
miss = df.isnull().sum() # Calculate the number of missing values in each column
print("Missing Values:")
print(miss)
```

# Convert Date Column

```python
# Convert to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Extract useful time features
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["MonthName"] = df["Date"].dt.month_name()
df["Day"] = df["Date"].dt.day
df["Weekday"] = df["Date"].dt.day_name()

# Display first few rows
df.head()
```

# Extract City

```python
df["City"] = df["Branch"].str.split("-").str[0].str.strip()

df[["Branch", "City"]].head()
```

# Recalculate Total Amount

```python
df["TotalAmount"] = df["Quantity"] * df["UnitPrice"] # recalculate the "TotalAmount" column based on "Quantity" and "UnitPrice" to ensure accuracy

df.head() # display the first few rows of the DataFrame to verify the changes
```
